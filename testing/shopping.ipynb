```python
#find null values
for column in df.columns:
    print(f"{column} has {df[column].isnull().sum()} null values")
```

```python
df_shopping = df.dropna()
df_shopping
```

```python
#find duplicate values
print(f"duplicate values: {df_shopping.duplicated().sum()}")
```

```python
df_shopping = df_shopping.drop('CustomerID', axis=1)
```

```python
df_shopping
```

```python
def change_string(member):
    if member == "Yes":
        return 1
    else:
        return 0
    
    
df_shopping["Card Member"] = df_shopping["Card Member"].apply(change_string)
df_shopping
```

```python
df_shopping["Annual Income"] = df_shopping["Annual Income"] / 1000
df_shopping
```

```python
columns = ["Card_Member","Age","Annual_Income","Spending_Score"]
df_shopping.columns = columns
df_shopping
```

```python
df_shopping.to_csv("./resources/shopping_data_cleaned.csv", index=False)
```
